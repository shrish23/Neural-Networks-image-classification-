At its core, PyTorch is a library for processing tensors. A tensor is a number, vector, matrix or any n-dimensional array. Let's create a tensor with a single number:

```python
t1 = torch.tensor(4.)
t1
```

```python
t1.dtype
```

```python
# Vector
t2 = torch.tensor([1.,2,3,4])
t2
```

```python
# Matrix
t3 = torch.tensor([[5.,6],
                   [7,8],
                   [9,10]])
t3
```

## Tensor Operations and gradients

```python
# Create tensors
x = torch.tensor(3.)
w = torch.tensor(4., requires_grad=True)
b = torch.tensor(5., requires_grad=True)
x,w,b
```

```python
y = w*x+b
y
```

To caculate the gradient descent we have to calculate the derivative.
To calculate the derivative of y we can invoke .backward method

```python
# Compute the derivative
y.backward()
```

THe derivative of the y is stored in the .grad property of the respective tensors

```python
# Display gradients
print('dy/dx', x.grad)
print('dy/dw',w.grad)
print('dy/db',b.grad)
```

Note: x.grad is None because x doesn't have a require_grad set to True as we are not interested in the derivative of the x.

This is optimization as well as we do not want to calculate the derivative of every variable

# Interoperability with Numpy
Numpy is a popular open source library used for mathematical and scientific computing in Python. It enables efficient operations on large multi-dimensional arrays, and has a large ecosystem of supporting libraries:

Matplotlib for plotting and visualization
OpenCV for image and video processing
Pandas for file I/O and data analysis
Instead of reinventing the wheel, PyTorch interoperates really well with Numpy to leverage its existing ecosystem of tools and libraries.

Here's how we create an array in Numpy:

```python
import numpy as np

x = np.array([[1,2],
              [3,4]])
x
```

```python
y = torch.from_numpy(x)
y
```

```python
x.dtype, y.dtype
```

The interoperability between PyTorch and Numpy is really important because most datasets you'll work with will likely be read and preprocessed as Numpy arrays.
